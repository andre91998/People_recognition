```python
import cv2
import numpy as np
import time
```

```python
# Video Capture 
#capture = cv2.VideoCapture(0)
capture = cv2.VideoCapture("/Users/Andre/Documents/GitHub/pedestrian_recognition/method4/pedestrian.mp4")


# Keeps track of what frame we're on
frameCount = 0
Frame=[]

while(1):
    # Return Value and the current frame
    ret, frame = capture.read()

    #  Check if a current frame actually exist
    if not ret:
        break

    frameCount += 1
    # Resize the frame
    grayFrame = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    resizedFrame = cv2.resize(grayFrame, (0, 0), fx=1.50, fy=1.50)
    
    dim=np.shape(resizedFrame)
    high=dim[0]
    lenght=dim[1]
    # Create an array in which I keep track of previous frames
    Frame.append(resizedFrame)
    if frameCount>5:
        A=Frame[1:]
        Frame=A
        
    images = np.stack(Frame, axis=2)
    median=np.median(images,axis=2,keepdims=(high ,lenght))
    image=median[:,:,0]
    
    
    img=(image-resizedFrame)**2/256
    img_int=img.astype(np.uint8)
    img_int[img_int<=10]=0
    
    modified_frame=np.where(img_int > 0, resizedFrame, 0)
    
    
    

    cv2.imshow('Frame', modified_frame)
    #cv2.imshow('median',image)

    time.sleep(0.01)
    k = cv2.waitKey(1) & 0xff
    if k == 27:
        break

capture.release()
cv2.destroyAllWindows()
```
